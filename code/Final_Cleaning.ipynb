```python
# noticing there are commas that will interfere with modeling, we will remove them and ensure all data is in integer form
```

```python
cols = ['Total_Avg_$',
'Total_Med_$',
'Total_Employment',
'Total_Mem_%',
'Construction_Avg_$',
'Construction_Med_$',
'Construction_Employment',
'Construction_Mem_%',
'Total_Fatalities',
'Construction_Fatalities']
def rm_c(df):
    for c in cols:
        df[c].replace(',', '', regex=True, inplace=True)
        df[c] = df[c].astype(int)
    return df
```

```python
# to standardize column names across all tables, the years will be removed from the poverty columns
final_2010 = rm_c(final_2010.rename(columns={'2010_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2011 = rm_c(final_2011.rename(columns={'2011_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2012 = rm_c(final_2012.rename(columns={'2012_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2013 = rm_c(final_2013.rename(columns={'2013_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2014 = rm_c(final_2014.rename(columns={'2014_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2015 = rm_c(final_2015.rename(columns={'2015_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2016 = rm_c(final_2016.rename(columns={'2016_Pct_Below_Pov':'Pct_Below_Pov'}))
final_2017 = rm_c(final_2017.rename(columns={'2017_Pct_Below_Pov':'Pct_Below_Pov'}))
```
