```python
poverty15 = pd.read_csv('../data/ACS_15.csv', header=1, index_col='Geography')
poverty15.head()
```

Given the excessive amount of columns (most of which are not relevant to our project) in the poverty dataframes, we will build a function to transpose the dataframe, remove these columns by index, and then reset the dataframe.

```python
def fix2(df):
    df = df.T.iloc[2:8] # tranpose to remove with row locator
    df = df.T # revert to original
    df.columns = ['Total Population', 'Population MOE','Population Below Poverty', # rename remaining columns
                  'Poverty MOE','Percentage Below Poverty','Percentage MOE']
    df.index = df.index.rename('State') # rename index column
    return df.head()
```

```python
poverty2015 = fix2(poverty15)
poverty2016 = fix2(poverty16)
poverty2017 = fix2(poverty17)
```

```python
poverty2015.to_csv('../data/poverty2015.csv')
poverty2016.to_csv('../data/poverty2016.csv')
poverty2017.to_csv('../data/poverty2017.csv')
```

The MOE columns represent margin of error for each estimate column. While these are important to understanding the data, they are not necessary to our modeling, and will therefore be removed in the 'cleaned' csv;however, they will still be available in the original saved csv in our 'data' folder.

```python
def drop(df):
    df.drop(columns=['Population MOE', 'Poverty MOE','Percentage MOE'],inplace=True)
    return df.head(2)
```

```python
# running the function on dfs for 2015, 2016, and 2017
drop(poverty2015)
poverty2015.to_csv('../data/cleaned_poverty2015.csv')
drop(poverty2016)
poverty2016.to_csv('../data/cleaned_poverty2016.csv')
drop(poverty2017)
poverty2017.to_csv('../data/cleaned_poverty2017.csv')
```
